### cleaning Numerical columns with charactors 

<code> remove text unit and convert to numbers</code>

```python
df['Size'] = pd.to_numeric(
    df['Size']
    .astype(str)                                     # ← astype
    .str.replace(r'[^0-9.]', '', regex=True)         # ← regex
    .replace('', None),
    errors='coerce'
)
```

```python
df
```

<code>removing the '+' for better comparison </code>

```python
df['Installs'] = pd.to_numeric(
    df['Installs']
    .astype(str)                                     # ← astype
    .str.replace(r'[^0-9.]', '', regex=True)         # ← regex
    .replace('', None),
    errors='coerce'
)
```

<code>now both Size and Installs are float type Data</code>

```python
df
```

```python
df.info()
```

<code>changing "price" column from str to float

```python
df['Price'] = pd.to_numeric(
    df['Price']
    .astype(str)                                     # ← astype
    .str.replace(r'[^0-9.]', '', regex=True)         # ← regex
    .replace('', None),
    errors='coerce'
)
```

```python
df
```

```python
df.info()
```

```python
df['Price'].value_counts()
```

<code>parse last_updated as datetime</code>

* changing the last updated coulumn from string to float

```python
df["Last Updated"] = pd.to_datetime(df["Last Updated"], errors='coerce')
```

```python
def convert_size(size):
    size = str(size).strip().upper()    # normalize everything to uppercase
    if size.endswith('M'):
        return float(size[:-1])
    elif size.endswith('K'):            # handles both 'k' and 'K'
        return float(size[:-1]) / 1024
    else:
        return None

df['Size_MB'] = df['Size'].apply(convert_size)

# Verify
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))      # Set the figure size to 4x4 inches
sns.histplot(data=df_new,x='Rating',bins=20,color="yellow",hue="Type")    # Create a histogram of the 'Rating' column,bins=(bar quanity)
plt.title('Distribution of Ratings') 
plt.xlabel("Rating")     # Set the x-axis label of the plot
plt.ylabel("Frequency")   # Set the y-axis label of the plot
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.histplot(data=df_new,x='Installs',bins=20,color="blue",hue="Type")
plt.title('Distribution of Installs')
plt.xlabel("No of Installation")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.histplot(data=df_new,x='Reviews',bins=20,color="#65F903")
plt.title('Distribution of Reviews')
plt.xlabel("No of Reviews")
plt.ylabel("Frequency")
plt.show()
```
