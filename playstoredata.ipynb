* ### Here the "Size" column is in str format, converting them into integer/float by removing the M and K for better numerical analysis

```python
import numpy as np  # ← add this

def clean_size(value):
    if pd.isna(value) or str(value).strip() == 'Varies with device':
        return np.nan

    value = str(value).strip()

    if value.endswith('M'):
        return float(value[:-1])

    elif value.endswith('k'):
        return float(value[:-1]) / 1024

    else:
        return np.nan

df['size_mb'] = df['Size'].apply(clean_size)
#* takes messy app size values and converts them into clean numeric values in megabytes so your data becomes usable
```
